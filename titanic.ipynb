First of all let's install the most basic of the libaraies that will heelp us in manipulating the data . 

```python
import pandas as pd 
import numpy as np 
```

Now let's try to see how our data looks like . Now the pd.read_csv('File_name') is basically a method via which you can read any csv file . It loosley translates into pandas read this csv file ('this is the file name ') and the df is basically dataframe but in simpleterms its you file . So you canname it anything but df is widely used as the calling name for dataframe. 

```python
df  = pd.read_csv('train.csv')
df
```

Now i saw a few NaN values they are basiclly non_existing value or empty spaces . so let's see how many of them are in total . so we call in another function df.is_na()  now this will return a whole dataframe of true and false for if a value is missing or not (try for yourself and see the magic) . Now it still won't help so we call in another function on them .sum() now this is a magical function that return the aggregate of all the nan values . now .sum() doesn't only work for is_na() but it return aggregate for a lot of method . like val_counts() and other . 

```python
df.isna().sum()
```

Now we will start our EDA . so we will make our data clean fill in missing values and then make the dataframe look good . 

```python
mean_age = df['Age'].mean()
mean_age
```

```python
df['Age'] = df['Age'].fillna(mean_age)
```

```python
df['Age'].isna().sum()
```

```python
cabin_mean = df['Cabin'].mode()[0]
cabin_mean
```

```python
df['Cabin'] = df['Cabin'].fillna(cabin_mean)
df['Cabin'].isna().sum()
```

```python
embarked_mode = df['Embarked'].mode()[0]
embarked_mode
```

```python
df['Embarked'] = df['Embarked'].fillna(embarked_mode)
df['Embarked'].isna().sum()
```

```python
df['p_class_3'] = pd.get_dummies(df['Pclass'])[3].astype(int)
df
```

```python
df['Embarked_Q'] = pd.get_dummies(df['Embarked'])['Q'].astype(int)
df
```

```python
df.drop(['Pclass', 'Embarked' , 'Name' , 'Cabin' , 'Ticket'] , axis = 1 , inplace = True)
df
```

```python
df['Sex'] = df['Sex'].map({'male' : 0 , 'female' : 1})
```

```python
df
```

```python
max_age = df['Fare'].max()
df['Fare'] = df['Fare']/max_age
df
```

Finally done now let's import the train_test_split() from sklearn . 

```python
from sklearn.model_selection import train_test_split 
```

Now let's split up our data in 80-20 . 

```python
X = df.drop('Survived' , axis = 1)
y = df['Survived']

X_train , y_train , X_test , y_test = train_test_split(X,y,random_state=42 , test_size=0.2)
```

Let's import our model from sklearn . 

```python
from sklearn.linear_model import LogisticRegression 
```

```python
model = LogisticRegression(max_iter = 1000)
```

Let's train our model 

```python
model.fit(X_train , y_train)
```

Now let's predict our result for our test . 

```python
y_pred = model.predict(X_test)
```

Now let's test how accurate our model is against the orginl data 

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
model_accuracy  = accuracy_score(y_test , y_pred)
print('Acurracy =' , model_accuracy)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

80% is not bad . nice start . now same steps are applied for Xgboost try it for yourself .
